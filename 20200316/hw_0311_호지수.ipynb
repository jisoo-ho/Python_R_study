#### 1. 다음은 학생 성적 관리 프로그램으로 리스트와 딕셔너리로 표현하고 for문을 사용해서 학생별 총점과 평균을 구했다. 딕셔너리로 객체를 각각 만드니 번거롭고 실수가 발생될 가능성이 높았다. Student 클래스를 만들고 각 학생 객체로 구성되는 학생 리스트를 작성하여 for문을 이용하여 출력하는 방식으로 프로그램을 개선 하세요.

```python
import pandas as pd
#1번 방법
class Student:
    lis = [[87, 98, 88, 95], [92, 98, 96, 98], [76, 96, 94, 90], [98, 92, 96, 92], [95, 98, 98, 98], [64, 88, 92, 92] ]
    name = ['조인성', '연하진', '구지연', '나선주', '윤아린', '윤영월']
    for i in range(6):
        print('이름:', name[i],'총점:',sum(lis[i]),'평균:',sum(lis[i])/len(lis[i]))

print('1번 방법 \n')
#2번 방법
students = [
    { "name": "조인성", "korean": 87, "math": 98, "english": 88, "science": 95 },
    { "name": "연하진", "korean": 92, "math": 98, "english": 96, "science": 98 },
    { "name": "구지연", "korean": 76, "math": 96, "english": 94, "science": 90 },
    { "name": "나선주", "korean": 98, "math": 92, "english": 96, "science": 92 },
    { "name": "윤아린", "korean": 95, "math": 98, "english": 98, "science": 98 },
    { "name": "윤명월", "korean": 64, "math": 88, "english": 92, "science": 92 }
]

class Student:
    def __init__(self, name):
        self.name=name
    
    def getScore(self, *args):
        self.kor=args[0]
        self.mat=args[1]
        self.eng=args[2]
        self.sci=args[3]

alist = list()

for stu in students:
    a = Student(stu['name'])
    a.getScore(stu['korean'],stu['math'],stu['english'],stu['science'])
    alist.append(a)

for i in range(len(alist)):
    total = alist[i].kor+alist[i].mat+alist[i].eng+alist[i].sci
    print('이름:',alist[i].name,'총점:',total,'평균',total/4)
print('2번 방법')
```

#### 2. 원의 둘레와 넓이를 구하는 메소드를 포함하고 있는 Circle 클래스를 활용해서 반지름이 10인 원의 둘레와 넓이를 구하세요. 

```python
PI = 3.141592
half = 10
class Circle:
    global PI
    global half
    def dul(self, half):
        return PI * (half*2)
    def nulb(self, half):
        return PI * (half**2)
    
c = Circle()
print('둘레:',c.dul(half))
print('넓이:',c.nulb(half))
```

#### 3. 랜덤하게 1000명의 키와 몸무게를 파일 쓰기모드로 작성 후 저장하세요.

```python
import random
f=open('tallandweight.txt','w')
for i in range(1,1001):
    tall = '%d 번의 키는 %d cm 이고 ' % (i,random.randint(40,100))
    weight = '몸무게는 %d kg 입니다. \n' % random.randint(140,200)
    f.write(tall)
    f.write(weight)
f.close()
```

#### 5. 재귀함수를 이용해 다음 팩토리얼을 출력하세요.
1!: \
2!: \
3!: \
4!: \
5!: 

```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n-1)
    
print('1!:', factorial(1))
print('2!:', factorial(2))
print('3!:', factorial(3))
print('4!:', factorial(4))
print('5!:', factorial(5))
```
